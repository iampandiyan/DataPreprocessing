```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from Univariate import Univariate
univariateObj=Univariate()
dataset=pd.read_csv("kidney_disease.csv")
dataset
```

```python
dataset.isna().sum()
```

```python
quan,qual=univariateObj.quanQual(dataset)
```

```python
quan
```

```python
qual
```

```python
dataset[quan].isna().sum()
```

```python
dataset[qual].isna().sum()
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(dataset[quan])
quandatset=imp_mean.transform(dataset[quan])
quandataFrame=pd.DataFrame(quandatset, columns =quan)
quandataFrame
```

```python
quandataFrame.isna().sum()
```

```python
imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mode.fit(dataset[qual])
qualdataset=imp_mode.transform(dataset[qual])
qualdataFrame=pd.DataFrame(qualdataset, columns =qual)
qualdataFrame
```

```python
qualdataFrame.isna().sum()
```

```python
#Concat 2 dataframe horizondally
preProcessedKidnydataset = pd.concat([quandataFrame, qualdataFrame], axis=1)
preProcessedKidnydataset
```

```python
preProcessedKidnydataset.isna().sum()
```

```python
preProcessedKidnydataset.isnull().sum()
```

```python
preProcessedKidnydataset.to_csv('PreprocessedKidnyDataset.csv', index=False)
```
